```python
from bs4 import BeautifulSoup
import requests
import itertools
import pandas as pd
```

```python
page_link = "https://fr.wikipedia.org/wiki/Liste_des_communes_de_France_les_plus_peupl%C3%A9es"
page_response = requests.get(page_link).text
soup = BeautifulSoup(page_response, 'lxml')
```

```python
Commune = []
content = soup.find("table",{"class":"wikitable sortable"})
for row in content.find_all('tr'):
    ville = row.find('a').get("title")
    Commune.append(ville)
del(Commune[0])
```

```python
top10 = Commune[0:10]
```

```python
top10
```

```python
def get_distance_between(origin, destination):
    url = f"https://fr.distance24.org/route.json?stops={origin}|{destination}"
    data = requests.get(url).json()
    return data['distance']
```

```python
combinations = list(itertools.combinations(top10, 2))
```

```python
distances = []
for origin, dest in combinations:
    distance = get_distance_between(origin, dest)
    distances.append(distance)
```

```python
df = pd.DataFrame(combinations, columns=['origin', 'dest'])
```

```python
df['distance'] = distances
```
